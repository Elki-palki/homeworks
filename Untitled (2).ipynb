# Функция двух переменных

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(-5, 5, 0.0001)
y = np.arange(-5, 5, 0.0001)
```

```python
f1 = 3 * np.square(x) + 5 * np.square(y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, f1, label='График функции двух переменных')
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
x, y = np.meshgrid(x, y)
f1 = 3 * np.square(x) + 5 * np.square(y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f1, cmap='viridis')
plt.title('Поверхность')
```

```python
def df(x, y):
    return np.array([6*x, 10 * y])
```

```python
x0, y0 = 5, 10
lambda_ = 0.1
eps = math.e**(-10)
history = [(x0, y0)]

while True:
    x_next, y_next = (x0 - lambda_ * df(x0, y0)[0], y0 - lambda_ * df(x0, y0)[1])
    if (x_next - x0) ** 2 + (y_next - y0) ** 2 < eps ** 2:
        break
    x0, y0 = x_next, y_next
    history.append((x0, y0))
```
